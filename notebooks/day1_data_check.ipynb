# Day 1: Live NAV Data Ingestion and Validation

```python
import requests
import pandas as pd
from pathlib import Path

# API URL given by the company
hdfc_url = "https://api.mfapi.in/mf/125497"

# Send GET request to API
response = requests.get(hdfc_url)

# Check whether request was successful
print("Status code:", response.status_code)

# Convert JSON response to Python dictionary
hdfc_json = response.json()

# Show scheme information
print("\nScheme information:")
print(hdfc_json["meta"])
```

```python
# Convert NAV history from JSON into a DataFrame
hdfc_nav_df = pd.DataFrame(hdfc_json["data"])

# Rename columns to clear names
hdfc_nav_df.columns = ["date", "nav"]

# Add scheme details
hdfc_nav_df["amfi_code"] = hdfc_json["meta"]["scheme_code"]
hdfc_nav_df["scheme_name"] = hdfc_json["meta"]["scheme_name"]

# Convert NAV to number and date to datetime
hdfc_nav_df["nav"] = pd.to_numeric(hdfc_nav_df["nav"], errors="coerce")
hdfc_nav_df["date"] = pd.to_datetime(hdfc_nav_df["date"], format="%d-%m-%Y", errors="coerce")

# Save as a raw CSV file
output_file = Path("../data/raw/hdfc_top_100_direct_live_nav.csv")
hdfc_nav_df.to_csv(output_file, index=False)

print("Saved file:", output_file)
print("Shape:", hdfc_nav_df.shape)

display(hdfc_nav_df.head())
```

```python
schemes = {
    "sbi_bluechip": 119551,
    "icici_bluechip": 120503,
    "nippon_large_cap": 118632,
    "axis_bluechip": 119092,
    "kotak_bluechip": 120841
}

all_live_nav = []

for short_name, scheme_code in schemes.items():
    url = f"https://api.mfapi.in/mf/{scheme_code}"

    try:
        response = requests.get(url)
        response.raise_for_status()

        scheme_json = response.json()

        # Convert API NAV records to DataFrame
        temp_df = pd.DataFrame(scheme_json["data"])
        temp_df.columns = ["date", "nav"]

        # Add metadata
        temp_df["amfi_code"] = scheme_json["meta"]["scheme_code"]
        temp_df["scheme_name"] = scheme_json["meta"]["scheme_name"]
        temp_df["short_name"] = short_name

        # Clean types
        temp_df["nav"] = pd.to_numeric(temp_df["nav"], errors="coerce")
        temp_df["date"] = pd.to_datetime(temp_df["date"], format="%d-%m-%Y", errors="coerce")

        # Save each scheme separately as raw CSV
        file_path = Path(f"../data/raw/{short_name}_live_nav.csv")
        temp_df.to_csv(file_path, index=False)

        all_live_nav.append(temp_df)

        print(f"✓ Downloaded {short_name}: {temp_df.shape}")
        print("  Saved:", file_path)

    except Exception as error:
        print(f"✗ Could not download {short_name}")
        print("  Error:", error)
```

```python
combined_live_nav_df = pd.concat(all_live_nav, ignore_index=True)

combined_file = Path("../data/raw/five_key_schemes_live_nav.csv")
combined_live_nav_df.to_csv(combined_file, index=False)

print("Combined file saved:", combined_file)
print("Combined shape:", combined_live_nav_df.shape)

display(combined_live_nav_df.head())
```

```python
import pandas as pd
from pathlib import Path

# Go from notebooks folder to data/raw folder
data_folder = Path("../data/raw")

# Find all CSV files
csv_files = sorted(data_folder.glob("*.csv"))

# Create the datasets dictionary
datasets = {}

for file in csv_files:
    # Skip the API files you downloaded later
    if "live_nav" in file.name.lower() or "five_key_schemes" in file.name.lower():
        continue

    df = pd.read_csv(file)
    datasets[file.stem] = df
    print("Loaded:", file.name, "| Shape:", df.shape)

print("\nDataset names available:")
print(list(datasets.keys()))
```

```python
# Get both datasets directly from the datasets dictionary
fund_master = datasets["01_fund_master"]
nav_history = datasets["02_nav_history"]

# Extract unique AMFI codes from both datasets
fund_master_codes = set(fund_master["amfi_code"].unique())
nav_history_codes = set(nav_history["amfi_code"].unique())

# Compare codes
missing_in_nav_history = fund_master_codes - nav_history_codes
extra_in_nav_history = nav_history_codes - fund_master_codes

# Print results
print("Fund master AMFI codes:", len(fund_master_codes))
print("NAV history AMFI codes:", len(nav_history_codes))

print("\nCodes in fund_master but NOT in nav_history:")
print(missing_in_nav_history)

print("\nCodes in nav_history but NOT in fund_master:")
print(extra_in_nav_history)

# Final validation result
if len(missing_in_nav_history) == 0:
    print("\n✓ Validation passed: Every AMFI code in fund_master exists in nav_history.")
else:
    print("\n⚠ Validation issue: Some fund_master AMFI codes are missing from nav_history.")
```
